```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Mine', 'Rock'], yticklabels=['Mine', 'Rock'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt


le = LabelEncoder()
y_test_numerical = le.fit_transform(y_test)
y_pred_numerical = le.transform(y_pred)


fpr, tpr, thresholds = roc_curve(y_test_numerical, y_pred_numerical)
roc_auc = auc(fpr, tpr)


plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np


y_test_numerical = le.transform(y_test)
y_pred_numerical = le.transform(y_pred)


mse = mean_squared_error(y_test_numerical, y_pred_numerical)
print(f'Mean Squared Error (MSE): {mse}')


mae = mean_absolute_error(y_test_numerical, y_pred_numerical)
print(f'Mean Absolute Error (MAE): {mae}')

rmse = np.sqrt(mse)
print(f'Root Mean Squared Error (RMSE): {rmse}')
```
